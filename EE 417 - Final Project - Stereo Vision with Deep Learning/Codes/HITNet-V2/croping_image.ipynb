```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
from glob import glob

# Path to ground truth disparity maps
GT_DIR = "/content/drive/MyDrive/EE 417/data-scene-flow/training/disp_occ_0"
gt_files = sorted(glob(os.path.join(GT_DIR, "*.png")))

def find_crop_boundary(gt_path):
    disp = cv2.imread(gt_path, cv2.IMREAD_UNCHANGED).astype(np.float32) / 256.0
    valid_mask = disp > 0

    # Calculate percentage of blank pixels per row
    row_blanks = np.mean(~valid_mask, axis=1) * 100

    # Find last row index where 100% of pixels are blank
    cutoff_row = 0
    for i, val in enumerate(row_blanks):
        if val == 100.0:
            cutoff_row = i
        else:
            break  # stop at first non-blank row

    return cutoff_row, row_blanks

# Analyze a sample image
sample_idx = 0
cutoff, blanks = find_crop_boundary(gt_files[sample_idx])
print(f"Cutoff row for cropping: {cutoff}")

# Optional: visualize for confirmation
disp = cv2.imread(gt_files[sample_idx], cv2.IMREAD_UNCHANGED).astype(np.float32) / 256.0
plt.imshow(disp[cutoff:, :], cmap='turbo')
plt.title(f'Cropped Disparity Map from row {cutoff}')
plt.show()
```

```python
import os
import cv2
import numpy as np
from glob import glob
from tqdm import tqdm

# ========== Define Paths ==========
LEFT_DIR  = "/content/drive/MyDrive/EE 417/data-scene-flow/training/image_2"
RIGHT_DIR = "/content/drive/MyDrive/EE 417/data-scene-flow/training/image_3"
GT_DIR    = "/content/drive/MyDrive/EE 417/data-scene-flow/training/disp_occ_0"

# Output directories
LEFT_OUT  = LEFT_DIR.replace("image_2", "image_2_cropped")
RIGHT_OUT = RIGHT_DIR.replace("image_3", "image_3_cropped")
GT_OUT    = GT_DIR.replace("disp_occ_0", "disp_occ_0_cropped")

os.makedirs(LEFT_OUT, exist_ok=True)
os.makedirs(RIGHT_OUT, exist_ok=True)
os.makedirs(GT_OUT, exist_ok=True)

# ========== Find Global Cutoff ==========
def get_cutoff_row(gt_path):
    disp = cv2.imread(gt_path, cv2.IMREAD_UNCHANGED).astype(np.float32) / 256.0
    valid_mask = disp > 0
    row_blanks = np.mean(~valid_mask, axis=1) * 100
    for i, val in enumerate(row_blanks):
        if val < 100.0:
            return i
    return len(row_blanks)

gt_files = sorted(glob(os.path.join(GT_DIR, "*.png")))
cutoff_rows = [get_cutoff_row(p) for p in tqdm(gt_files, desc="Finding cutoff rows")]
GLOBAL_CUTOFF = max(cutoff_rows)
print(f"✅ Global cutoff row to safely crop all images: {GLOBAL_CUTOFF}")

# ========== Prepare Common Filenames ==========
left_files  = {os.path.basename(p): p for p in glob(os.path.join(LEFT_DIR, "*.png"))}
right_files = {os.path.basename(p): p for p in glob(os.path.join(RIGHT_DIR, "*.png"))}
gt_files    = {os.path.basename(p): p for p in glob(os.path.join(GT_DIR, "*.png"))}

common_filenames = sorted(set(left_files) & set(right_files) & set(gt_files))
print(f"✅ Found {len(common_filenames)} files present in all three folders.")

# ========== Crop & Save ==========
for fname in tqdm(common_filenames, desc="Cropping and saving"):
    l_img = cv2.imread(left_files[fname])
    r_img = cv2.imread(right_files[fname])
    g_img = cv2.imread(gt_files[fname], cv2.IMREAD_UNCHANGED)

    l_crop = l_img[GLOBAL_CUTOFF:, :]
    r_crop = r_img[GLOBAL_CUTOFF:, :]
    g_crop = g_img[GLOBAL_CUTOFF:, :]

    cv2.imwrite(os.path.join(LEFT_OUT, fname), l_crop)
    cv2.imwrite(os.path.join(RIGHT_OUT, fname), r_crop)
    cv2.imwrite(os.path.join(GT_OUT, fname), g_crop)

print("🎉 Done! Cropped and saved aligned images.")
```

```python
import os
import cv2
import matplotlib.pyplot as plt

# Cropped folders
LEFT_OUT  = "/content/drive/MyDrive/EE 417/data-scene-flow/training/image_2_cropped"
RIGHT_OUT = "/content/drive/MyDrive/EE 417/data-scene-flow/training/image_3_cropped"
GT_OUT    = "/content/drive/MyDrive/EE 417/data-scene-flow/training/disp_occ_0_cropped"

# Check sample files
sample_filenames = sorted(os.listdir(LEFT_OUT))[:3]

for fname in sample_filenames:
    left_path  = os.path.join(LEFT_OUT, fname)
    right_path = os.path.join(RIGHT_OUT, fname)
    gt_path    = os.path.join(GT_OUT, fname)

    # Load images
    left_img  = cv2.imread(left_path)[..., ::-1]  # BGR to RGB
    right_img = cv2.imread(right_path)[..., ::-1]
    gt_img    = cv2.imread(gt_path, cv2.IMREAD_UNCHANGED).astype(float) / 256.0

    # Print file and shape info
    print(f"File: {fname}")
    print(f"  Left Image  shape: {left_img.shape}")
    print(f"  Right Image shape: {right_img.shape}")
    print(f"  GT Map      shape: {gt_img.shape}")

    # Plot
    plt.figure(figsize=(15, 4))
    plt.subplot(1, 3, 1)
    plt.imshow(left_img)
    plt.title('Left Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(right_img)
    plt.title('Right Image')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(gt_img, cmap='turbo')
    plt.title('Disparity GT')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
